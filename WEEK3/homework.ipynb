```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score

# 1. ë°ì´í„° ë¶ˆëŸ¬ì˜¤ê¸°
df = pd.read_csv('wine.csv')

# 2. í”¼ì²˜, ë¼ë²¨ ë¶„ë¦¬
X = df.drop('Wine', axis=1)
y = df['Wine'] - 1  # 0ë¶€í„° ì‹œì‘í•˜ê²Œ ì¡°ì •

# 3. ì •ê·œí™”
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 4. train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y
)

# 5. ë¶„ë¥˜ê¸° ì„ íƒ & í•™ìŠµ (Random Forest)
clf = RandomForestClassifier(n_estimators=100, random_state=42)
clf.fit(X_train, y_train)

# 6. ì˜ˆì¸¡ ë° í‰ê°€
y_pred = clf.predict(X_test)
print("âœ… Accuracy:", accuracy_score(y_test, y_pred))
print("\nğŸ“Š Classification Report:\n", classification_report(y_test, y_pred))
```

```python
from tensorflow.keras.layers import Input, Dense, Concatenate
from tensorflow.keras.models import Model
from tensorflow.keras.utils import to_categorical
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd

# ë°ì´í„° ë¡œë”© ë° ì „ì²˜ë¦¬
df = pd.read_csv('wine.csv')
X = df.drop('Wine', axis=1)
y = df['Wine'] - 1

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y
)

y_train_cat = to_categorical(y_train)
y_test_cat = to_categorical(y_test)

# ë”¥ëŸ¬ë‹ ì…ë ¥ ì •ì˜
input_layer = Input(shape=(X_train.shape[1],))

# Wide part (ì„ í˜•)
wide = Dense(3, activation='linear')(input_layer)

# Deep part (ì€ë‹‰ì¸µ)
deep = Dense(64, activation='relu')(input_layer)
deep = Dense(32, activation='relu')(deep)
deep = Dense(16, activation='relu')(deep)

# Wide + Deep ê²°í•©
combined = Concatenate()([wide, deep])
output = Dense(3, activation='softmax')(combined)

model = Model(inputs=input_layer, outputs=output)

# ì»´íŒŒì¼
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# í•™ìŠµ
history = model.fit(X_train, y_train_cat,
                    epochs=50,
                    batch_size=16,
                    validation_split=0.2,
                    verbose=1)
```
